```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import pandas as pd

# https://scitools.org.uk/cartopy/docs/v0.16/gallery/hurricane_katrina.html#sphx-glr-gallery-hurricane-katrina-py
# https://stackoverflow.com/questions/53290602/how-to-use-cartopy-to-create-colored-us-states
```

```python
stations = pd.read_csv("../Data/weather/stations.tsv", delimiter="\t")
lats = stations["LATITUDE"].to_list()
lons = stations["LONGITUDE"].to_list()
```

```python
map_extents = [-128, -64, 23, 46]

fig = plt.figure(figsize=(12, 10))
geo_ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
geo_ax.set_extent(map_extents)

geo_ax.stock_img()
geo_ax.coastlines()
geo_ax.add_feature(cfeature.STATES)
geo_ax.add_feature(cfeature.LAKES)
geo_ax.add_feature(cfeature.RIVERS)

geo_ax.scatter(lons, lats)
```
